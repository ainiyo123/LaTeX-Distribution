## Handle errors in TinyTeX
- When you compile a document to PDF through LaTeX, you may run into errors like these:

```python
! LaTeX Error: File `ocgbase.sty' not found.

!pdfTeX error: pdflatex (file 8r.enc):
  cannot open encoding file for reading

!pdfTeX error: /usr/local/bin/pdflatex (file tcrm0700):
  Font tcrm0700 at 600 not found
```

- To fix this is very simple using the tinytex::parse_install(). This function helps you quickly install the missing packages in your LaTeX distribution:

```python
# if the log file is filename.log
tinytex::parse_install("filename.log")

# or use the `text` argument
tinytex::parse_install(
  text = "! LaTeX Error: File `ocgbase.sty' not found."
)
# it will install the "ocgx2" package
```

- Note that a similar function tinytex::parse_packages() can also find missing packages in your LaTeX distribution even if you are not using TinyTeX:

```python
# if the log file is filename.log
tinytex::parse_packages("filename.log")

# or use the `text` argument
tinytex::parse_packages(
  text = "! LaTeX Error: File `ocgbase.sty' not found."
)
# it should return "ocgx2"
```

## PDF via LaTeX

1) After you compile the pdf file successfully, you should be able to use the pdf via LaTeX option in Jupyter Notebook. Make sure to compile in the console first since you have to install the missing packages first in order for the LaTeX extension working correctly in Jupyter Notebook.

2) A common issue here is that you may get a error message like this:

```python
nbconvert failed: pdflatex not found on PATH
```

- This error is caused due to the uncorrected PATH variable in the environment.
- For this error, there are two ways to fix it:

1. Open your command line prompt and enter:

```python
export PATH=$PATH=/Library/TeX/texbin
```

2. Or you can add this to your kernel:

```python
"env": {"PATH":"$PATH"}
```

3) Then your PDF via LaTeX option would work perfectly.
